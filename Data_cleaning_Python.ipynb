# Data Cleaning using Python

```python
import pandas as pd
```

```python
# url of file uploaded in my github
# by doing this we don't need to upload data file everytime we run this colab file

url = 'https://github.com/MekhzZ/Data-Cleaning-Challenge-in-Excel-SQL-Python/raw/refs/heads/main/Customer%20Call%20List.xlsx'
```

```python
df = pd.read_excel(url)
```

```python
df.head()
```

```python
# first let's remove duplicates

df.drop_duplicates(inplace=True)
df
```

```python
# deleting 'Not_Useful_Column'

df.drop('Not_Useful_Column', axis=1, inplace=True)
```

```python
df
```

```python
# formatting 'Last_Name'

df['Last_Name'] = df['Last_Name'].str.strip('/. ')
```

```python
df['Last_Name'].head()
```

```python
# replacing 'Yes' with Y and No with N

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y').str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y').str.replace('No', 'N')
```

```python
df
```

```python
# let's fill na with ' '

df = df.replace('N/a', ' ')
df = df.replace('NaN', ' ')
df = df.fillna('')
```

```python
df
```

```python
# let's create columns wrt 'Address'

df[['Street_address', 'State', 'Zip_code']] = df['Address'].str.split(',', expand=True)
df
```

```python
# now let's delete 'Address' as we extracted required addresses
# also let's delete 'Zip_coe' column as we have only one value in it

df.drop(['Address', 'Zip_code'], axis=1, inplace=True)
```

```python
df
```

```python
df.fillna('', inplace = True)
```

```python
df
```

```python
# now let's clean 'Phone_Number' : it's messy and not well formatted

def format_num(num):
  num = str(num)
  if '-' in num:
    return num
  if '|' in num:
    return num.replace('|', '-')
  if '/' in num:
    return num.replace('/', '-')
  return f"{num[:3]}-{num[3:6]}-{num[6:]}"
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(format_num)
df
```

```python
# we have '--' remaining in 'Phone_Number'

df['Phone_Number'] = df['Phone_Number'].str.replace('--', '')
df
```

```python
# now we have most columns formatted
# let's extract data of people who wants to get contact i.e: 'Do_Not_Contact' = N

for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] == 'Y':
    df.drop(x, inplace=True)

df
```

```python
# again removed blank spaces

for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] == '':
    df.drop(x, inplace=True)

df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
# let's remove the one with no any Phone_Number

df = df[df['Phone_Number'] != '']
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df.to_excel('cleaned_data.xlsx', index=False)

# output will be in runtime files
```
